```python
def save_tensor(tensor, filename, rownames=None, colnames=None):
    # 检查 tensor 是 PyTorch 张量、Pandas Series 还是 Numpy 数组
    if isinstance(tensor, torch.Tensor):
        # 如果是 PyTorch 张量，先移至 CPU，再转换为 NumPy 数组
        data = tensor.cpu().numpy()
    elif isinstance(tensor, pd.Series):
        # 如果是 Pandas Series，使用 .values 属性
        data = tensor.values
    elif isinstance(tensor, np.ndarray):
        # 如果是 Numpy 数组，直接使用
        data = tensor
    else:
        raise ValueError("Unsupported data type for saving.")

    # 转换为Pandas DataFrame
    df = pd.DataFrame(data)

    # 设置行和列名称
    if rownames is not None:
        df.index = rownames
    if colnames is not None:
        df.columns = colnames

    # 将DataFrame保存为CSV文件
    df.to_csv(filename, header=True, index=True)
```
